```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import yfinance as yf
```

# Treasury Bond Futures

```python
bd = yf.download('ZB=F', progress=False) # 30 years bond
bd2 = yf.download('ZT=F', progress=False) # 2 years note
bd3 = yf.download('ZN=F', progress=False) # 10 years note
bd2
```

```python
def FOMC ():
    
    plt.axvline(pd.to_datetime('2022-09-22'), color='r', linestyle='--')
    plt.axvline(pd.to_datetime('2022-11-03'), color='r', linestyle='--')
    plt.axvline(pd.to_datetime('2022-12-15'), color='b', linestyle='--')
    plt.axvline(pd.to_datetime('2023-02-02'), color='b', linestyle='--')

    plt.axvline(pd.to_datetime('2023-03-23'), color='b', linestyle='--')
    plt.axvline(pd.to_datetime('2023-05-04'), color='b', linestyle='--')
    plt.axvline(pd.to_datetime('2023-06-15'), color='k', linestyle='--')


plt.figure(figsize=(10,12))

plt.subplot(3,1,1)
plt.title('30y Bond Futures')
plt.plot(bd.iloc[-230:].Close)
FOMC()

plt.subplot(3,1,2)
plt.plot(bd.iloc[-230:].Volume)
plt.xticks(rotation=30)
FOMC()

std1 = bd.Close.pct_change().rolling(10).std().iloc[-230:].apply(lambda x : x*np.sqrt(252))
std2 = bd.Close.pct_change().rolling(20).std().iloc[-230:].apply(lambda x : x*np.sqrt(252))
std3 = bd.Close.pct_change().rolling(30).std().iloc[-230:].apply(lambda x : x*np.sqrt(252))

plt.subplot(3,1,3)
plt.plot(std1)
plt.plot(std2)
plt.plot(std3)
plt.xticks(rotation=30)
FOMC()

plt.show()
```
