```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import yfinance as yf
```

USD/KRW

```python
usdkrw = yf.download('KRW=X', progress=False) # 30 years bond

usdkrw.head()
```

```python
usdkrw.iloc[-280:].Close.plot()
```

```python
upper = usdkrw.iloc[-280:].Close.rolling(7).mean() + usdkrw.iloc[-280:].Close.rolling(7).std() * 1.96
lower = usdkrw.iloc[-280:].Close.rolling(7).mean() - usdkrw.iloc[-280:].Close.rolling(7).std() * 2.16
ma = usdkrw.iloc[-280:].Close.rolling(7).mean()

plt.figure(figsize=(10,4))
plt.plot(usdkrw.iloc[-280:].Close)
plt.plot(upper, alpha=0.6)
plt.plot(lower, alpha=0.6)
plt.plot(ma)

print(upper.iloc[-1])
print(lower.iloc[-1])
```
